```python
# 음료수 얼려 먹기 복습

from collections import deque

n, m = map(int, input().split())

graph = []

for i in range(n):
    graph.append(list(map(int,input())))

def bfs(x,y):
    
    queue = deque()
    queue.append((x,y))
    
    dx = [-1,1,0,0]
    dy = [0,0,-1,1]
    
    graph[x][y] = 1

    
    while queue:
        x, y = queue.popleft((x,y))
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx >= n or ny >= m or nx < 0 or ny < 0:
                continue
            
            if graph[nx][ny] == 0:
                graph[nx][ny] = 1
                queue.append((nx,ny))
                

        
count = 0
for i in range(n):
    for j in range(m):
        if graph[i][j] == 0:
            count += 1
            bfs(i,j)
print(count)
```
